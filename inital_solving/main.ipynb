```python
# QP
def runMQP(G,fn):

    print(len(G.nodes),len(G.edges))
    NWeight = [G.nodes[i]['NodeWeight'] for i in G.nodes]
#     numOriginNodes=47824
#     assert(sum(NWeight) == numOriginNodes)
    EWeight = [G.edges[i]['EdgeWeight'] for i in G.edges]
    from datetime import date
    td = str(date.today())

    nowg = G
    node_list = list(nowg.nodes)
    N = len(node_list)
    # N = node_list
    # E = list(nowg.edges)
    # L = nx.laplacian_matrix(nowg)
    W = nx.adj_matrix(nowg,weight='EdgeWeight')

#     assert(np.sum(W.data)==cut*2)
    Wary = W.toarray() 
    linesum = np.sum(Wary,0)
    print(linesum)
    maxline = linesum.max() / 2 
#     maxline=0
    # Wp = Wary + np.eye(Wary.shape[0]) * cut * 2
    num_machines = 12

    #Model
    model = gp.Model('Graph Partition')

    #variables
    # map_nodes = model.addVars(ij, vtype=GRB.BINARY, name="partition")
    map_nodes = model.addMVar((num_machines,N),lb=0.0, ub=1.0,vtype=GRB.BINARY, name="partition")


    max_band = model.addMVar(1,name="max_band")

    #constraint0: each node on one machine only
    certainty = model.addConstrs((map_nodes[:,i].sum() == 1 
                                  for i in range(N)), name='certainty')
    #constraint1: CPU etc.
    cap_cpu = 1000
    machine_capacity = model.addConstrs((map_nodes[i,:] @ np.array(NWeight)  <= cap_cpu 
                                        for i in range(num_machines)), name='machine_capacity')
    #constraint2: NIC 
    # print(map_nodes.select(0, '*'))

    # for machine1 in range(num_machines):
    #     nic = model.addMQConstr(L, np.array([-1]), '<', 0, map_nodes.select(machine, '*'), map_nodes.select(machine, '*'), max_band)
    #constraint2: Edge
    from itertools import combinations
    for i,j in combinations(range(num_machines),2):
        psd = map_nodes[i,:] @ W @ map_nodes[j,:] + maxline*sum(map_nodes[k, :] @ map_nodes[k, :] for k in [i,j])
        edge = model.addConstr(psd - max_band <= maxline*(map_nodes[i,:].sum()+map_nodes[j].sum()))
    #     edge = model.addConstr(map_nodes[i,:] @ W @ map_nodes[j,:] <= max_band[0])

    #constraint3: breaking symmetry

    # desym1 = model.addConstrs((map_nodes[q,i] - map_nodes[q,q] <= 0
    #                           for q in range(num_machines) for i in range(q+1,len(node_list))) , name="smaller")

    # desym2 = model.addConstrs((map_nodes[q,i] == 0 
    #                            for q in range(num_machines) for i in range(q)), name='former0')
    # desym1 = model.addConstrs((gp.quicksum(map_nodes[i,j] for j in N) <= gp.quicksum(map_nodes[i+1,j] for j in N)
    #                                      for i in range(num_machines-1)), name='desymmetry')

    # machineonly5 = model.addConstr(gp.quicksum(map_nodes[i,i] for i in range(num_machines)) <= 5)
    #Objective
    model.setObjective(max_band,GRB.MINIMIZE)

    # model.setParam('ConcurrentMIP',2)
    model.setParam('Threads', 12)
    model.setParam('NoRelHeurTime', 300)
    model.setParam('TimeLimit', 600)
    model.setParam('LogFile', logdir+fn+td+"MQPgurobi.log")
    model.setParam('SolFiles', "MQP/" + fn+td)
    # model.setParam('MIPFocus', 3)

    # Verify model formulation

#     model.write(logdir+fn+td+'MQP.lp')
    # mp = model.presolve()
    # mp.write(logdir+td+'presolved-convex.lp')
    # Run optimization engine

    model.optimize()
```

```python
from datetime import date

td = str(date.today())
N = range(len(node_list))
# N = node_list
# E = list(nowg.edges)
L = nx.laplacian_matrix(nowg)
W = nx.adj_matrix(nowg)
Wary = W.toarray() 
num_machines = 5

list_ij = []
for i in range(num_machines):
    for j in N:
        tp = i,j
        list_ij.append(tp)
ij = gp.tuplelist(list_ij)

#Model
model = gp.Model('Graph Partition')

#variables
map_nodes = model.addVars(ij, lb=0, ub=1, name="partition")
max_band = model.addMVar(1, name="max_band")

#constraint0: each node on one machine only
certainty = model.addConstrs((gp.quicksum(map_nodes[i,j] for i in range(num_machines)) >= 1
                              for j in N), name='certainty')
#constraint1: CPU etc.
cap_cpu = 1000
machine_capacity = model.addConstrs((gp.quicksum(map_nodes[i,j] for j in N) <= cap_cpu 
                                     for i in range(num_machines)), name='machine_capacity')
#constraint2: NIC 
# print(map_nodes.select(0, '*'))

# for machine1 in range(num_machines):
#     nic = model.addMQConstr(L, np.array([-1]), '<', 0, map_nodes.select(machine, '*'), map_nodes.select(machine, '*'), max_band)

#constraint2: Edge
from itertools import combinations
comb = combinations(range(num_machines),2)
for i,j in comb:
    edge = model.addMQConstr(W, np.array([-1]), GRB.LESS_EQUAL, 0, map_nodes.select(i, '*'), 
                             map_nodes.select(j, '*'), max_band)
#     edge = model.addQConstr(gp.quicksum(W * map_nodes.select(i,'*') * map_nodes.select(j,'*')) <= max_band)


#constraint3: breaking symmetry
desym1 = model.addConstrs((map_nodes[q,i] - map_nodes[q,q] <= 0
                          for q in range(num_machines) for i in range(q+1,len(node_list))) , name="smaller")

desym2 = model.addConstrs((map_nodes[q,i] == 0 
                           for q in range(num_machines) for i in range(q)), name='former0')


# desym1 = model.addConstrs((gp.quicksum(map_nodes[i,j] for j in N) <= gp.quicksum(map_nodes[i+1,j] for j in N)
#                                      for i in range(num_machines-1)), name='desymmetry')

# machineonly5 = model.addConstr(gp.quicksum(map_nodes[i,i] for i in range(num_machines)) <= 5)
#Objective


model.setObjective(max_band,GRB.MINIMIZE)

# model.setParam('ConcurrentMIP',2)
model.setParam('Threads', 120)
model.setParam('NonConvex', 2)
model.setParam('NoRelHeurTime', 5*24*3600)
model.setParam('TimeLimit', 10*24*3600)
model.setParam('LogFile', logdir+td+"lp-gurobi.log")
model.setParam('SolFiles', "LP-solutions/" + td+ "mymodel")
# model.setParam('MIPFocus', 3)

# Verify model formulation

model.write(logdir+'lp-GraphPartition.lp')

# Run optimization engine

model.optimize()
```

```python
# node size proportional to degrees, note that this uses matplotlib instead of graphviz
# plt_width = 0.4
def draw_igraph_with_plt(g,name):
    thisg = nx.Graph(g.get_edgelist())
#     thisg = g
#     pos = nx.multipartite_layout(thisg,'subset')
    pos = nx.spring_layout(thisg, seed=5)
    d=dict(thisg.degree)
    plt.figure(figsize=(10,10))
    # plt.xlim(-plt_width,plt_width)
    # plt.ylim(-plt_width,plt_width)
    nx.draw_networkx_nodes(thisg,pos,nodelist=d.keys(), node_size=[15*np.sqrt(v) for v in d.values()], \
                           node_color=[v * 1 for v in d.values()], cmap=plt.cm.Blues,vmin=0, \
                           vmax=2*max(list(d.values())), edgecolors='black')
    nx.draw_networkx_edges(thisg,pos,alpha=0.2,width=0.1)
    plt.savefig(pngdir+"{}.pdf".format(name))
#     plt.savefig(pngdir+"test.png")

def global_transitivity_igraph(g):
    return g.transitivity_undirected('zero')

def draw_coreness_igraphs(graph_dict : dict, fill : int, filename : str) -> None :
    coreness_fig, coreness_ax = plt.subplots(figsize=(7,7))
    
    for i in graph_dict:
        g = graph_dict[i]
        corenesslist = g.coreness()
        corenesslist.sort()
        list_to_show = [0]*(fill-len(corenesslist)) + corenesslist
        assert(len(list_to_show)==2000)
        coreness_ax.plot(np.arange(2000),list_to_show, label=i)
        
    coreness_ax.set_title('Coreness for all nodes')
    coreness_ax.set_xlabel('Node index')
    coreness_ax.set_ylabel('Coreness')
    coreness_ax.legend()
    coreness_fig.savefig(pngdir+"{}.pdf".format(filename))
    
def draw_coreness_cdf_igraphs(graph_dict : dict, filename : str) -> None :
    coreness_fig, coreness_ax = plt.subplots(figsize=(21,7))
    
    all_CY = []
    all_label = []
    max_tail = 0
    for i in graph_dict:
        g = graph_dict[i]
#         corenesslist = g.coreness()
        corenesslist = g.shell_index()
        max_tail = max(max_tail,max(corenesslist))
        from collections import Counter
        coredict = dict(Counter(corenesslist))
        Y = np.array([coredict[i] if i in coredict else 0 for i in range(max_tail+1)])
        CY = np.cumsum(Y / Y.sum())
        all_CY.append(CY)
        all_label.append(i)
#         coreness_ax.plot(np.arange(tail+1),CY, label=i)
    
    for i in range(len(all_CY)):
        coreness_ax.plot(np.arange(max_tail+1),
                         np.append(all_CY[i], np.ones(max_tail + 1 - np.shape(all_CY[i])[0])), label=all_label[i])
    coreness_ax.set_title('Coreness CDF')
    coreness_ax.set_xlabel('Coreness')
    coreness_ax.legend()
#     coreness_fig.savefig(pngdir+"{}.pdf".format(filename))
    
```

# Modified K-L

```python
# modified K-L algorithm source code
from itertools import count
from networkx.utils import not_implemented_for, py_random_state, BinaryHeap
from networkx.algorithms.community.community_utils import is_partition

__all__ = ["kernighan_lin_bisection"]


def _kernighan_lin_sweep(edges, side):
    """
    This is a modified form of Kernighan-Lin, which moves single nodes at a
    time, alternating between sides to keep the bisection balanced.  We keep
    two min-heaps of swap costs to make optimal-next-move selection fast.
    """
    costs0, costs1 = costs = BinaryHeap(), BinaryHeap()
    for u, side_u, edges_u in zip(count(), side, edges):
        cost_u = sum(w if side[v] else -w for v, w in edges_u)
        costs[side_u].insert(u, cost_u if side_u else -cost_u)

    def _update_costs(costs_x, x):
        for y, w in edges[x]:
            costs_y = costs[side[y]]
            cost_y = costs_y.get(y)
            if cost_y is not None:
                cost_y += 2 * (-w if costs_x is costs_y else w)
                costs_y.insert(y, cost_y, True)

    i = totcost = 0
    while costs0 and costs1:
        u, cost_u = costs0.pop()
        _update_costs(costs0, u)
        v, cost_v = costs1.pop()
        _update_costs(costs1, v)
        totcost += cost_u + cost_v
        yield totcost, i, (u, v)



def kernighan_lin_bisection(G, partition=None, max_iter=10, weight="weight", seed=None):
    """Partition a graph into two blocks using the Kernighan–Lin
    algorithm.

    This algorithm partitions a network into two sets by iteratively
    swapping pairs of nodes to reduce the edge cut between the two sets.  The
    pairs are chosen according to a modified form of Kernighan-Lin, which
    moves node individually, alternating between sides to keep the bisection
    balanced.

    Parameters
    ----------
    G : graph

    partition : tuple
        Pair of iterables containing an initial partition. If not
        specified, a random balanced partition is used.

    max_iter : int
        Maximum number of times to attempt swaps to find an
        improvemement before giving up.

    weight : key
        Edge data key to use as weight. If None, the weights are all
        set to one.

    seed : integer, random_state, or None (default)
        Indicator of random number generation state.
        See :ref:`Randomness<randomness>`.
        Only used if partition is None

    Returns
    -------
    partition : tuple
        A pair of sets of nodes representing the bipartition.

    Raises
    -------
    NetworkXError
        If partition is not a valid partition of the nodes of the graph.

    References
    ----------
    .. [1] Kernighan, B. W.; Lin, Shen (1970).
       "An efficient heuristic procedure for partitioning graphs."
       *Bell Systems Technical Journal* 49: 291--307.
       Oxford University Press 2011.

    """
    n = len(G)
    labels = list(G)
#     seed.shuffle(labels)
    index = {v: i for i, v in enumerate(labels)}

    if partition is None:
        side = [0] * (n // 2) + [1] * ((n + 1) // 2)
    else:
        try:
            A, B = partition
        except (TypeError, ValueError) as e:
            raise nx.NetworkXError("partition must be two sets") from e
        if not is_partition(G, (A, B)):
            raise nx.NetworkXError("partition invalid")
        side = [0] * n
        for a in A:
            side[index[a]] = 1

    if G.is_multigraph():
        edges = [
            [
                (index[u], sum(e.get(weight, 1) for e in d.values()))
                for u, d in G[v].items()
            ]
            for v in labels
        ]
    else:
        edges = [
            [(index[u], e.get(weight, 1)) for u, e in G[v].items()] for v in labels
        ]

    for i in range(max_iter):
        costs = list(_kernighan_lin_sweep(edges, side))
        min_cost, min_i, _ = min(costs)
        if min_cost >= 0:
            break

        for _, _, (u, v) in costs[: min_i + 1]:
            side[u] = 1
            side[v] = 0

    A = {u for u, s in zip(labels, side) if s == 0}
    B = {u for u, s in zip(labels, side) if s == 1}
    return A, B

num = 12000
thisg = sqrtg[num]
randomnodes = list(thisg)
np.random.shuffle(randomnodes)
# part1 = randomnodes[:6000]
# part2 = randomnodes[6000:10000]

kl = kernighan_lin_bisection(thisg,max_iter=5000)
len(thisg.edges)-len(thisg.subgraph(kl[0]).edges)-len(thisg.subgraph(kl[1]).edges)

# save partition to txt
np.savetxt(txtdir+"cluster6000-sqrtp10000.txt",np.array([i for i in kl[1]]),fmt='%s')
np.savetxt(txtdir+"cluster4000-sqrtp10000.txt",np.array([i for i in kl[0]]),fmt='%s')
np.savetxt(txtdir+"all-sqrtp10000.txt",p_random10000,fmt='%s')
```

```python
# log binning plot

import numpy as np
from collections import Counter
from math import log10
def drop_zeros(a_list):
    return [i for i in a_list if i>0]

def log_binning(counter_dict,bin_count=35):

    max_x = log10(max(list(counter_dict.keys())))
    max_y = log10(max(list(counter_dict.values())))
    max_base = max([max_x,max_y])

    min_x = log10(min(drop_zeros(list(counter_dict.keys()))))

    bins = np.logspace(min_x,max_base,num=bin_count)

    # Based off of: http://stackoverflow.com/questions/6163334/binning-data-in-python-with-scipy-numpy
    bin_means_y = (np.histogram(list(counter_dict.keys()),bins,weights=list(counter_dict.values()))[0] / np.histogram(list(counter_dict.keys()),bins)[0])
    bin_means_x = (np.histogram(list(counter_dict.keys()),bins,weights=list(counter_dict.keys()))[0] / np.histogram(list(counter_dict.keys()),bins)[0])

    return bin_means_x,bin_means_y


ba_g = g
ba_c = nx.degree_centrality(ba_g)

# To convert normalized degrees to raw degrees
ba_c = {k:int(v*(len(ba_g)-1)) for k,v in ba_c.items()}
ba_c2 = dict(Counter(ba_c.values()))

ba_x,ba_y = log_binning(ba_c2,50)

plt.xscale('log')
plt.yscale('log')
plt.scatter(ba_x,ba_y,c='r',marker='s',s=50)
# plt.scatter(ba_c2.keys(),ba_c2.values(),c='b',marker='x')
# plt.xlim((1e-4,1e-1))
# plt.ylim((.9,1e4))
plt.xlabel('Degrees')
plt.ylabel('Number of Nodes')
plt.title('Log-binning of degree distribution')
plt.show()
```
